```python
import mne
import numpy as np
import matplotlib.pyplot as plt

raw=mne.io.read_raw_fif('D:\Dataset\P01-raw.fif')
x=raw.get_data('Fp1')
x=x.reshape(-1,)
print(x.shape)
'''
NFFT : fft size  ->    
Fs : sampling frequency  ->    (   )
overlap :  fft block   
'''
fig, ax = plt.subplots(2,2,figsize=(16,8))
ax[0,0].specgram(x,NFFT=32768,Fs=65536,noverlap=int(16384*0.5))
ax[0,0].set_ylim([0,50])
ax[0,0].set_ylabel("Frequency(Hz)")
ax[0,0].set_title("FFT size : 32768, sampling frequency : 65536 Hz, overlap : 1/4")

ax[0,1].specgram(x,NFFT=32768,Fs=32768,noverlap=int(16384*0.5))
ax[0,1].set_ylim([0,50])
ax[0,1].set_ylabel("Frequency(Hz)")
ax[0,1].set_title("FFT size : 32768, sampling frequency : 32768 Hz, overlap : 1/4")

ax[1,0].specgram(x,NFFT=32768,Fs=65536,noverlap=int(2048*0.5))
ax[1,0].set_ylim([0,50])
ax[1,0].set_ylabel("Frequency(Hz)")
ax[1,0].set_title("FFT size : 32768, sampling frequency : 65536 Hz, overlap : 1/32")

ax[1,1].specgram(x,NFFT=4096,Fs=8192,noverlap=int(2048*0.5))
ax[1,1].set_ylim([0,50])
ax[1,1].set_ylabel("Frequency(Hz)")
ax[1,1].set_title("FFT size : 4096, sampling frequency : 8192 Hz, overlap : 1/4")
plt.show()
```
