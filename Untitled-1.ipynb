```python
import mne
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal
raw=mne.io.read_raw_fif('D:\Dataset\P01-raw.fif')
x=raw.get_data('Fp1')
raw.plot()
```

```python
sf=256.
time=np.arange(x.size)/sf

x=x.reshape(2478166,1)

fig, ax = plt.subplots(1,1,figsize=(12,4))
plt.plot(x)
plt.xlabel('time')
plt.ylabel('voltage')
plt.title('Fp1 voltage per second')
plt.show()

```

```python
win=sf*4
fs = 100
x=x.reshape(-1,)
freqs,psd=signal.welch(x,fs,nperseg=win)


psd=psd.reshape(-1,)

plt.figure(figsize=(8,4))
plt.plot(freqs,psd)
plt.xlabel('Frequency(Hz)')
plt.ylabel('PSD(V^2/Hz)')
plt.ylim([0,psd.max()*1.1])
plt.title("Welch's periodogram")
plt.xlim([0,freqs.max()])
plt.show()
```

```python
f,s=signal.periodogram(x,fs=216,scaling='density')
plt.figure(figsize=(40,10))
plt.semilogy(f,s)
plt.xlim(0,50)
plt.ylim(1e-19,1e-5)
plt.show()
```

```python
def integral(a,b):
    mask=((f>a) & (f<b))
    s_new=s[mask]
    return s_new.sum()/s_new.shape[0]

print('delta wave result : ',integral(0.2,3.99))
print('theta wave result : ',integral(4,7.99))
print('alpha wave result : ',integral(8,12.99))
print('beta wave result : ',integral(13,30))
print('gamma wave result : ',integral(30,50))
```
