```python
from scipy.fft import fft,ifft
import mne
import numpy as np
import matplotlib.pyplot as plt


raw=mne.io.read_raw_fif('D:\Dataset\P01-raw.fif')
x=raw.get_data('Fp1')
x=x.reshape(-1,1)

def psi(T,f0=6):
    x=np.linspace(-2*np.pi,2*np.pi,T)
    return (np.pi**-0.25)*np.exp(1j*f0*x-x**2/2)

def wavelet_convolution(tup):
    f=tup[0]
    T=tup[1]
    f_len=np.shape(f)[0]
    f_hat=np.append(f,np.zeros(T))
    h=psi(T)
    h_hat=np.append(h,np.zeros(f_len))

    return ifft(fft(f_hat)*fft(h_hat))[round(T/2):round(T/2)+f_len]

def cwt(f,t0):
    f_len=np.shape(f)[0]
    result=wavelet_convolution([x,t0])
    return result


```

```python
fig, ax = plt.subplots(2,2,figsize=(12,10))
ax[0,0].plot(psi(10))
ax[0,0].set_title('wavelet fuction(10Hz)')
ax[0,1].plot(psi(20))
ax[0,1].set_title('wavelet fuction(20Hz)')
ax[1,0].plot(psi(30))
ax[1,0].set_title('wavelet fuction(30Hz)')
ax[1,1].plot(psi(40))
ax[1,1].set_title('wavelet fuction(40Hz)')
plt.show()
```

```python
time=np.arange(x.size)/100

fig, ax = plt.subplots(3,2,figsize=(12,16))
ax[0,0].plot(time,x)
ax[0,0].set_title('raw data')
ax[0,1].plot(time,cwt(x,10))
ax[0,1].set_title('10Hz convolution result')
ax[1,0].plot(time,cwt(x,20))
ax[1,0].set_title('20Hz convolution result')
ax[1,1].plot(time,cwt(x,30))
ax[1,1].set_title('30Hz convolution result')
ax[2,0].plot(time,cwt(x,40))
ax[2,0].set_title('40Hz convolution result')
ax[2,1].plot(time,cwt(x,50))
ax[2,1].set_title('50Hz convolution result')
plt.show()
```

```python
import pywt

def plot_wavelet(ax, time, signal, scales, waveletname = 'gaus4', 
                 cmap = plt.cm.seismic, title = '', ylabel = '', xlabel = ''):
    
    [coefficients, frequencies] = pywt.cwt(signal, scales, waveletname)
    power = (abs(coefficients)) ** 2
    plt.imshow(abs(coefficients),cmap='copper',extent=[0,10,50,2],
               interpolation='bilinear',aspect='auto',vmax=abs(coefficients).max(),vmin=-abs(coefficients).min())
    plt.gca().invert_yaxis()
    plt.show()

scales = np.arange(1, 50)
title = 'Wavelet Transform (Power Spectrum) of signal'
ylabel = 'Period (years)'
xlabel = 'Time'
time=np.arange(x.size)/100

fig, ax = plt.subplots(figsize=(10, 10))
plot_wavelet(ax, time, x, scales, xlabel=xlabel, ylabel=ylabel, title=title)
plt.show()
```
